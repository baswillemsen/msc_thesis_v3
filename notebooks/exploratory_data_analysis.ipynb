# Stationarity

```python
timeframe = 'm'
df = read_data(get_data_path(timeframe=timeframe), file_name=f'total_{timeframe}')
```

```python
var_name = 'unempl'
fig_size = (10, 4)
log = True
diff_level = 1

# log
if log:
    df_target = np.log(df[df['country']=='france'].set_index('date')[var_name])
else:
    df_target = df[df['country']=='france'].set_index('date')[var_name]
df_target.index = pd.to_datetime(df_target.index)

fig, ax = plt.subplots(figsize=fig_size)
ax.plot(df_target)

ax.xaxis.set_major_locator(mdates.YearLocator())
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
if log:
    ax.set_title(f'Natural log of {var_name_formal[var_name]}')
else:
    ax.set_title(f'{var_name_formal[var_name]}')
ax.set_xlabel('Year')
ax.set_ylabel(f'{var_name_formal[var_name]}')

if log:
    plt.savefig(f'{var_name}_france_log.png', bbox_inches='tight', dpi=300)
else:
    plt.savefig(f'{var_name}_france.png', bbox_inches='tight', dpi=300)
plt.show()


if diff_level > 0:
    # stationary
    if log:
        df_target = np.log(df[df['country']=='france'].set_index('date')[var_name]).diff(diff_level)
    else:
        df_target = df[df['country']=='france'].set_index('date')[var_name].diff(diff_level)
    df_target.index = pd.to_datetime(df_target.index)

    fig, ax = plt.subplots(figsize=fig_size)
    ax.plot(df_target)

    ax.xaxis.set_major_locator(mdates.YearLocator())
    for tick in ax.get_xticklabels():
        tick.set_rotation(45)

    if log:
        ax.set_title(f'{diff_level}-m diff of natural log of {var_name_formal[var_name]}')
    else:
        ax.set_title(f'{diff_level}-m diff of {var_name_formal[var_name]}')
    ax.set_xlabel('Year')
    ax.set_ylabel(f'{var_name_formal[var_name]}')

    if log:
        plt.savefig(f'{var_name}_france_log_diff{diff_level}.png', bbox_inches='tight', dpi=300)
    else:
        plt.savefig(f'{var_name}_france_diff{diff_level}.png', bbox_inches='tight', dpi=300)
    plt.show()
```
